```python
#Import and load data

import os
import csv
import pandas as pd
import geopandas as gpd

cd = gpd.read_file("data/CD/lcd_000b21a_e.shp")
csd = gpd.read_file("data/CSD/lcsd000b21a_e.shp")
cma = gpd.read_file("data/CMA/lcma000b21a_e.shp")
```

```python
#Spatial join because no common column to make a standard join

sjoin = gpd.sjoin(csd , cma , how = 'left')
sjoin.head()
```

```python
#Shortening the CSDUID so that it can be joined with the CD table

sjoin['CSDUID_short'] = sjoin['CSDUID'].str[:4]
sjoin.head()
```

```python
#Joining the CD table

full_join = pd.merge(cd , sjoin , left_on = 'CDUID' , right_on = 'CSDUID_short')
full_join.head()
```

```python
#Filter census data within the GGH

ggh = full_join.loc[(full_join['CSDUID'].str.contains("3543"))|# Simcoe
                        (full_join['CSDUID'].str.contains("3522"))|# Dufferin
                        (full_join['CSDUID'].str.contains("3523"))|# Wellington
                        (full_join['CSDUID'].str.contains("3516"))|# Kawartha Lakes
                        (full_join['CSDUID'].str.contains("3515"))|# Peterborough
                        (full_join['CSDUID'].str.contains("3514"))|# Northumberland
                        (full_join['CSDUID'].str.contains("3518"))|# Durham
                        (full_join['CSDUID'].str.contains("3519"))|# York
                        (full_join['CSDUID'].str.contains("3520"))|# Toronto
                        (full_join['CSDUID'].str.contains("3521"))|# Peel
                        (full_join['CSDUID'].str.contains("3524"))|# Halton
                        (full_join['CSDUID'].str.contains("3530"))|# Waterloo
                        (full_join['CSDUID'].str.contains("3525"))|# Hamilton
                        (full_join['CSDUID'].str.contains("3529"))|# Brant
                        (full_join['CSDUID'].str.contains("3526"))|# Niagara
                        (full_join['CSDUID'].str.contains("3528"))]# Haldimand-Norlfolk
ggh
```

```python
# Subset the data for what is actually needed

ggh_sub = ggh[["CDNAME" , "CDUID" , "CSDNAME" , "CSDUID" , "CMANAME", "CMAUID" , "CMAPUID"]]
ggh_sub
```

```python
#export

ggh_sub.to_csv('ggh_boundaries.csv', index = False)
```
